```python
import numpy as np
import pandas as pd
% matplotlib inline
```

```python
power_example = []
cap = 10000
for i in range(1000):
    num = cap / 100000.0
    pcnt = np.random.power(a=num, size=1)[0]
    print("Cap: {0}, Num: {1}, Pct: {2}".format(cap, num, pcnt))
    cap = cap - cap * pcnt
    power_example.append(cap)
```

```python
df = pd.DataFrame()
df['points'] = power_example
```

```python
df.points.plot()
```
